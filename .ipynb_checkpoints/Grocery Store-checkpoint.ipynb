```python
def remove_redundant_rules(rules):
    non_redundant_rules = []
    for i in range(len(rules)):
        for j in range(i+1, len(rules)):
            if set(rules.iloc[i]['antecedents']).issubset(set(rules.iloc[j]['antecedents'])) and rules.iloc[i]['consequents'] == rules.iloc[j]['consequents']:
                break
        else:
            non_redundant_rules.append(rules.iloc[i])
    return pd.DataFrame(non_redundant_rules)

non_reduntant_rules = remove_redundant_rules(filtered_rules)
non_reduntant_rules_test = remove_redundant_rules(filtered_rules_test)
```

```python
def get_top_n_recommendations(user_items, rules, n=5):
    recommendations = {}

    for item in user_items:
        matching_rules = rules[rules['antecedents'].apply(lambda x: item in x)]
        for _, rule in matching_rules.iterrows():
            if item in rule['antecedents']:
                consequents = rule['consequents']
                confidence = rule['confidence']
                for consequent in consequents:
                    if consequent not in user_items:
                        if consequent not in recommendations:
                            recommendations[consequent] = confidence
                        else:
                            recommendations[consequent] = max(recommendations[consequent], confidence)

    # Sort recommendations by confidence and take the top n
    sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)[:n]
    return [item for item, _ in sorted_recommendations]

# Generate recommendations for each user in the test set
recommendations = {}
count = 0
for member_number, group in data_train.groupby('Member_number'):
    user_items = group['itemDescription'].tolist()
    recommendations[member_number] = get_top_n_recommendations(user_items, sorted_rules, n=5)
    count += 1

# Print recommendations
for member_number, items in recommendations.items():
    print(f"Recommendations for member {member_number}: {items}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Function to calculate precision, recall, and F1-score
def calculate_metrics(recommendations, ground_truth):
    precision_scores = []
    recall_scores = []
    f1_scores = []

    for member_number, recommended_items in recommendations.items():
        if member_number in ground_truth:
            true_items = ground_truth[member_number]

            y_true = [1 if item in true_items else 0 for item in recommended_items]
            y_pred = [1] * len(recommended_items)

            if sum(y_true) > 0:  # Only calculate if there are relevant items
                precision_scores.append(precision_score(y_true, y_pred, zero_division=0))
                recall_scores.append(recall_score(y_true, y_pred, zero_division=0))
                f1_scores.append(f1_score(y_true, y_pred, zero_division=0))

    avg_precision = sum(precision_scores) / len(precision_scores) if precision_scores else 0
    avg_recall = sum(recall_scores) / len(recall_scores) if recall_scores else 0
    avg_f1 = sum(f1_scores) / len(f1_scores) if f1_scores else 0

    return avg_precision, avg_recall, avg_f1

# Calculate and print the metrics for the first five members
avg_precision, avg_recall, avg_f1 = calculate_metrics(recommendations, ground_truth)

print(f"\nAverage Precision: {avg_precision:.4f}")
print(f"Average Recall: {avg_recall:.4f}")
print(f"Average F1-Score: {avg_f1:.4f}")
```

```python
def collaborative_filtering(user_item_matrix, threshold=0.1):
    # Calculate cosine similarity matrix
    similarity_matrix = cosine_similarity(user_item_matrix)
    
    # Convert to DataFrame
    similarity_df = pd.DataFrame(similarity_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)
    
    # Apply threshold to similarity matrix
    similarity_df[similarity_df < threshold] = 0
    
    return similarity_df


# Test collaborative filtering with adjusted similarity threshold with train data
similarity_df = collaborative_filtering(user_item_matrix, threshold=0.1)
```

```python
def item_based_collaborative_filtering(user_item_matrix, threshold=0.1):
    # Transpose the matrix to get item-user matrix
    item_user_matrix = user_item_matrix.T
    
    # Calculate cosine similarity matrix between items
    similarity_matrix = cosine_similarity(item_user_matrix)
    
    # Convert to DataFrame
    similarity_df = pd.DataFrame(similarity_matrix, index=item_user_matrix.index, columns=item_user_matrix.index)
    
    # Apply threshold to similarity matrix
    similarity_df[similarity_df < threshold] = 0
    
    return similarity_df
```

```python
# Load the dataset
data_train = pd.read_csv('Groceries data train.csv')
data_test = pd.read_csv('Groceries data test.csv')
```

```python
# Convert date to datetime format
data_train['Date'] = pd.to_datetime(data_train['Date'], dayfirst=True)

# Filter relevant columns
data_train = data_train[['Member_number', 'itemDescription']]

# Create a binary user-item matrix
train_user_item_matrix = data_train.pivot_table(index='Member_number', columns='itemDescription', aggfunc=lambda x: 1, fill_value=0)
print(train_user_item_matrix.head())

# Convert date to datetime format
data_test['Date'] = pd.to_datetime(data_test['Date'], dayfirst=True)

# Filter relevant columns
data_test = data_test[['Member_number', 'itemDescription']]

# Create a binary user-item matrix
test_user_item_matrix = data_test.pivot_table(index='Member_number', columns='itemDescription', aggfunc=lambda x: 1, fill_value=0)
print(test_user_item_matrix.head())
```

```python

# Recommendation Function
def recommend_items_adjusted(user_id, similarity_df, user_item_matrix, top_n=5):
    if user_id in user_item_matrix.index:
        user_similarity = similarity_df.loc[user_id].sort_values(ascending=False)[1:]
        user_items = user_item_matrix.loc[user_id]
        unseen_items = user_items[user_items == 0].index
        scores = {item: (user_item_matrix[item] * user_similarity).sum() for item in unseen_items}
        top_recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]
        return top_recommendations
    else:
        popular_items = user_item_matrix.sum().sort_values(ascending=False)[:top_n].index.tolist()
        return [(item, 0) for item in popular_items]
```

```python
def recommend_items_item_based(user_id, similarity_df, user_item_matrix, top_n=5):
    if user_id in user_item_matrix.index:
        user_items = user_item_matrix.loc[user_id]
        seen_items = user_items[user_items > 0].index
        
        scores = {}
        for item in seen_items:
            if item in similarity_df.columns:
                similar_items = similarity_df[item].sort_values(ascending=False)[1:]
                for similar_item, similarity in similar_items.items():
                    if similar_item not in seen_items:
                        if similar_item not in scores:
                            scores[similar_item] = similarity
                        else:
                            scores[similar_item] += similarity
            else:
                print(f"Item '{item}' not found in similarity matrix.")
        
        top_recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]
        
        return top_recommendations
    
    else:
        popular_items = user_item_matrix.sum().sort_values(ascending=False)[:top_n].index.tolist()
        return [(item, 0) for item in popular_items]
```

```python
# Generate Predictions
def generate_predictions(test_user_item_matrix, similarity_df, train_user_item_matrix, top_n=5):
    predictions = {}
    for user_id in test_user_item_matrix.index:
        recommendations = recommend_items_adjusted(user_id, similarity_df, train_user_item_matrix, top_n=top_n)
        predictions[user_id] = recommendations
    return predictions


# Generate predictions for the test set
def generate_item_based_predictions(test_user_item_matrix, similarity_df, train_user_item_matrix, top_n=5):
    predictions = {}
    for user_id in test_user_item_matrix.index:
        if user_id in train_user_item_matrix.index:  # Ensure the user exists in the training set
            recommendations = recommend_items_item_based(user_id, similarity_df, train_user_item_matrix, top_n=top_n)
            predictions[user_id] = recommendations
        else:
            predictions[user_id] = []  # No recommendations for users not in the training set
    return predictions

# Generate predictions
predictions = generate_predictions(test_user_item_matrix, similarity_df, train_user_item_matrix, top_n=5)
predictions
```

```python
# Generate predictions
predictions_item_based = generate_item_based_predictions(test_user_item_matrix, similarity_df_item, train_user_item_matrix, top_n=5)
predictions_item_based
```

```python
import numpy as np

def calculate_precision_recall(test_user_item_matrix, predictions, top_n=5):
    y_true = []
    y_pred = []

    for user_id in test_user_item_matrix.index:
        actual_items = test_user_item_matrix.loc[user_id]
        actual_bought_items = set(actual_items[actual_items > 0].index)

        recommended_items = [item for item, _ in predictions[user_id][:top_n]]
        predicted_recommended_items = set(recommended_items)

        tp = actual_bought_items.intersection(predicted_recommended_items)
        
        y_true.append(len(actual_bought_items))
        y_pred.append(len(predicted_recommended_items))
        
        precisions = len(tp) / len(predicted_recommended_items) if len(predicted_recommended_items) > 0 else 0
        recalls = len(tp) / len(actual_bought_items) if len(actual_bought_items) > 0 else 0

    average_precision = np.mean(precisions)
    average_recall = np.mean(recalls)
    
    return average_precision, average_recall


# Calculate and print precision and recall for the first set of predictions
precision, recall = calculate_precision_recall(test_user_item_matrix, predictions, top_n=5)
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')

# Calculate and print precision and recall for the second set of predictions (item-based)
precision_item, recall_item = calculate_precision_recall(test_user_item_matrix, predictions_item_based, top_n=5)
print(f'Precision: {precision_item:.4f}')
print(f'Recall: {recall_item:.4f}')
```

Collaborative Filtering

```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import mean_squared_error
from sklearn.metrics.pairwise import cosine_similarity


# Load the dataset
data_test = pd.read_csv('Groceries data train.csv')
data_train = pd.read_csv('Groceries data test.csv')
```

```python
# Concatenate the DataFrames
merged_data = pd.concat([data_train, data_test], axis=0, ignore_index=True)

# Reset the index (optional, to have a continuous index)
merged_data.reset_index(drop=True, inplace=True)

# Display the first few rows of the merged DataFrame
print(merged_data.head())
print(merged_data.tail())
```

```python
# Convert date to datetime format
merged_data['Date'] = pd.to_datetime(merged_data['Date'], dayfirst=True)

# Filter relevant columns
merged_data = merged_data[['Member_number', 'itemDescription']]

# Create a binary user-item matrix
user_item_matrix = merged_data.pivot_table(index='Member_number', columns='itemDescription', aggfunc=lambda x: 1, fill_value=0)
print(user_item_matrix.head())
```

```python
def collaborative_filtering(user_item_matrix, threshold=0.1):
    # Calculate cosine similarity matrix
    similarity_matrix = cosine_similarity(user_item_matrix)
    
    # Convert to DataFrame
    similarity_df = pd.DataFrame(similarity_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)
    
    return similarity_df


# Test collaborative filtering with adjusted similarity threshold with train data
similarity_df = collaborative_filtering(user_item_matrix, threshold=0.1)
```

```python
def recommend_items_adjusted(user_id, similarity_df, user_item_matrix, pattern_input, top_n=10):
    if user_id in user_item_matrix.index:
        # Get similarity scores for the given user
        user_similarity = similarity_df.loc[user_id].sort_values(ascending=False)[1:]  # Exclude self
        
        # Find items the user hasn't bought
        user_items = user_item_matrix.loc[user_id]
        unseen_items = user_items[user_items == 0].index
        
        # Calculate weighted sum of ratings for each unseen item using adjusted similarity scores
        scores = {}
        for item in unseen_items:
            item_similarity = user_item_matrix[item] * user_similarity
            scores[item] = item_similarity.sum()
        
        # Sort scores and recommend top N items
        top_recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]

        # Generate recommendations based on Apriori patterns
        apriori_recommendations = []
        for _, rule in pattern_input.iterrows():
            if set(rule['antecedents']).issubset(user_items[user_items == 1].index):
                apriori_recommendations.extend(rule['consequents'])
        
        apriori_recommendations = list(set(apriori_recommendations) - set(user_items[user_items == 1].index))
        apriori_recommendations = [item for item in apriori_recommendations if item not in dict(top_recommendations).keys()]  # Filtering items not in top_recommendations
        apriori_recommendations = apriori_recommendations[:top_n]
        
        # Combine Apriori and collaborative recommendations with source identifier
        combined_recommendations = [(item) for item in apriori_recommendations] + [(item) for item, _ in top_recommendations]
        
        return combined_recommendations
    
    else:
        # User is not in the training data, recommend popular items or use other methods
        popular_items = user_item_matrix.sum().sort_values(ascending=False)[:top_n].index.tolist()
        return [(item, 'Popular') for item in popular_items]

```

```python
# Load the dataset
data_train = pd.read_csv('Groceries data train.csv')
data_test = pd.read_csv('Groceries data test.csv')
# Convert date to datetime format
data_train['Date'] = pd.to_datetime(data_train['Date'], dayfirst=True)

# Filter relevant columns
data_train = data_train[['Member_number', 'itemDescription']]

# Create a binary user-item matrix
train_user_item_matrix = data_train.pivot_table(index='Member_number', columns='itemDescription', aggfunc=lambda x: 1, fill_value=0)
print(train_user_item_matrix.head())

# Convert date to datetime format
data_test['Date'] = pd.to_datetime(data_test['Date'], dayfirst=True)

# Filter relevant columns
data_test = data_test[['Member_number', 'itemDescription']]

# Create a binary user-item matrix
test_user_item_matrix = data_test.pivot_table(index='Member_number', columns='itemDescription', aggfunc=lambda x: 1, fill_value=0)
print(test_user_item_matrix.head())
```

```python
# Generate predictions for the test set
def generate_predictions(data_test, similarity_df, train_user_item_matrix,pattern_input, top_n=5):
    predictions = {}
    for user_id in data_test['Member_number']:
        recommendations = recommend_items_adjusted(user_id, similarity_df, train_user_item_matrix,pattern_input, top_n=top_n)
        predictions[user_id] = recommendations
    return predictions

# Generate predictions
predictions = generate_predictions(data_test, similarity_df, train_user_item_matrix,pattern_input, top_n=5)
predictions
```

```python
# Evaluate the recommendations
def evaluate_recommendations(test_user_item_matrix, predictions):
    precision_scores = []
    recall_scores = []
    f1_scores = []

    for user_id in predictions.keys():
        if user_id in test_user_item_matrix.index:
            actual_items = test_user_item_matrix.loc[user_id][test_user_item_matrix.loc[user_id] == 1].index.tolist()
            recommended_items = predictions[user_id]
            
            # Calculate precision, recall, and F1 score
            true_positives = len(set(recommended_items) & set(actual_items))
            false_positives = len(set(recommended_items) - set(actual_items))
            false_negatives = len(set(actual_items) - set(recommended_items))
            
            precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
            recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
            f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
            
            precision_scores.append(precision)
            recall_scores.append(recall)
            f1_scores.append(f1)

    avg_precision = np.mean(precision_scores)
    avg_recall = np.mean(recall_scores)
    avg_f1 = np.mean(f1_scores)

    return avg_precision, avg_recall, avg_f1

# Calculate evaluation metrics
precision, recall, f1 = evaluate_recommendations(test_user_item_matrix, predictions)
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```
